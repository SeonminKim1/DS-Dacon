## Daycon - 병원 개/폐업 분류 예측
- train.csv - 의료기관이 폐업했는지 여부를 포함하여 최근 2개년의 재무정보와 병원 기본정보 
- test.csv - 폐업 여부를 제외하고 train.csv와 동일 
- sample_submission.csv - inst_id와 open과 close를 예측하는 OC 두개의 열로 구성. OC의 값은 open 예측일 경우 1, close 예측일 경우 0.

## 변수 설명
### 병원정보
- inst_id - 각 파일에서의 병원 고유 번호
- OC – 영업/폐업 분류, 2018년 폐업은 2017년 폐업으로 간주함
- sido – 병원의 광역 지역 정보
- sgg – 병원의 시군구 자료
- openDate – 병원 설립일
- bedCount - 병원이 갖추고 있는 병상의 수
- instkind – 병원, 의원, 요양병원, 한의원, 종합병원 등 병원의 종류 
> 종합병원 : 입원환자 100명 이상 수용 가능 (general_hospital)
> 병원 : 입원 환자 30명 이상 100명 미만 수용 가능 (hospital)
> 의원 : 입원 환자 30명 이하 수용 가능 (clinic) 
> 한방 병원(한의원) : 침술과 한약으로 치료하는 의료 기관. (traditional_clinic)

```python
import os
import numpy as np
import pandas as pd

os.chdir('C:\\Users\\urse\\Desktop\\dataset\\Daycon-Hospital_Opening_closing_classification_prediction')

# 1. 데이터 로딩
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print('데이터 로딩 완료')
```

```python
# 2. 데이터 파악하기 - (2) dtypes 확인하기
test_df.dtypes
```

```python
# 2. 데이터 파악하기 - (3) 고유값 출력해보기 (dtypes - 수치형 변수들 제외하고)
print('############ train data ############ ')
for i in train_df.columns:
    if train_df[i].dtypes != np.float64 and train_df[i].dtypes != np.int64:
        print(i,'의 고유값들', train_df[i].unique(), '\n')

print('############ test data ############ ')
for i in test_df.columns:
    if test_df[i].dtypes != np.float64 and test_df[i].dtypes != np.int64:
        print(i,'의 고유값들', test_df[i].unique(), '\n')
        
        
# test_data는 employee 변수들이 float이아닌 object라는 것을 알 수 있음.
        
```

```python
# 3. 데이터 전처리 - (1) employee 변수들 object 변수에서 -> float &  OC close 변수 공백없애기.
train_df.employee1 = train_df.employee1.astype('float')
train_df.employee2 = train_df.employee2.astype('float')
train_df.OC = train_df.OC.astype('str').str.replace(' ', '') # close 데이터에 앞이 띄어져 있어서 공백이 있어서.

test_df.employee1 = test_df.employee1.astype('str').str.replace(',', '').astype('float')
test_df.employee2 = test_df.employee2.astype('str').str.replace(',', '').astype('float')

print('데이터 전처리 employee 변수들 object 변수에서 -> float &  OC close 변수 공백없애기 완료')
```

```python
# 4. 데이터 전처리 - (2) factor, numeric columns 구분
# Combining the train and test data
# 데이터 전처리 값 처리 등 더 편하기 위해서 잠깐 합침.
train_test_prod = train_df.append(test_df)

# Get the object and numeric columns seperately
# object 타입과 정수형 타입들의 columns들 골라내기
factor_columns = train_test_prod.select_dtypes(include=['object']).columns
numeric_columns = train_test_prod.columns.difference(factor_columns)

print('### factor_columns \n', factor_columns)
print('### numeric_columns \n', numeric_columns)
```

```python
# 5. 데이터 전처리 - (3) NA , NULL 값 등 처리

# 데이터 430번은 dental_clinic 인데 잘못 적혀 있었음.
# 데이터 413번은 Bedcount가 비어 있어서 채워줌.
train_test_prod.loc[train_test_prod.inst_id == 430, ['instkind']] = 'dental_clinic'
train_test_prod.loc[train_test_prod.inst_id == 430, ['bedCount']] = 0
train_test_prod.loc[train_test_prod.inst_id == 413, ['bedCount']] = -999

#Fill the empty values in the object columns as "Not sure"
# 범주형 변수들 NA를 Not sure 로 채움.
train_test_prod[factor_columns] = train_test_prod[factor_columns].fillna('Not_sure')

# 정수형 변수들 -999로 채워둠.
#Fill all the empty values in the numeric columns as -999
train_test_prod[numeric_columns] = train_test_prod[numeric_columns].fillna(-999)

print('데이터 전처리 - (3) NA , NULL 값 등 처리 완료')
```

```python
# Label Encoder 변환 전
train_test_prod.head()
```

```python
# 6. 데이터 전처리 - (4) Object columns LabelEncoder()로 변환하기

from sklearn.preprocessing import LabelEncoder

# 범주형 변수들 목록 : 'OC', 'sido', 'instkind', 'ownerChange'
# Convert all the object columns to numeric since the ML algorithms don't accept object features directly 
fac_le = LabelEncoder()
train_test_prod[factor_columns] = train_test_prod.loc[:,factor_columns].apply(lambda x : fac_le.fit_transform(x))
train_test_prod.head()
```

```python
# 7. 데이터 전처리 - (5) Object columns LabelEncoder()로 변환하기
# 다시 train_data, test_data 분리 
# Splitting back data to train prod and test prod
train_prod = train_test_prod.loc[train_test_prod.OC != 0,]
test_prod = train_test_prod.loc[train_test_prod.OC == 0,]
train_prod['OC'] = train_prod['OC'] - 1

print('train_df', train_prod.shape)
print('test_df', test_prod.shape)
```

```python
# 8. 데이터 전처리 - (6) 훈련 데이터 및 검증 데이터 생성

#Obtain the submission ID to create the submission file later
sub_id = test_prod.inst_id # 병원 ID

#Get the dependent and independent column
dep = 'OC' # 훈련 데이터 정답 라벨
no_dep = train_prod.columns.difference([dep])  # OC를 제외한 columns list 

# 훈련 데이터 / 훈련 데이터 정답 라벨 / 테스트 데이터 생성.
# OC 제외 데이터 / OC / OC 제외 데이터
train_prod_X = train_prod[no_dep]
train_prod_Y = train_prod[dep]
test_prod_X = test_prod[no_dep]

print("데이터 전처리 - (6) 훈련 데이터 및 검증 데이터 생성 완료")
```

```python
# 9. 데이터 학습 - (0) 라이브러리 import
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn import svm
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
import shap
import xgboost as xgb
import lightgbm as lgb
import seaborn as sns
import matplotlib.pyplot as plt

print('데이터 학습 - (0) 라이브러리 import 완료')
```

```python
############################################################################
############ Random Forest
############################################################################
# 10. 데이터 학습 - (1) Random Forest 변수 설정 및 학습 및 테스트

estimators = 10
np.random.seed(100)
RF_prod = RandomForestClassifier(n_estimators = estimators)

# 랜덤 포레스트 훈련 데이터 학습
RF_prod_model = RF_prod.fit(train_prod_X, train_prod_Y) 

# 랜덤 포레스트 테스트 데이터 테스트 
RF_prod_prediction = RF_prod.predict_proba(test_prod_X)[:,1] 

sub_RF = pd.DataFrame({'inst_id' : sub_id , 'OC' : RF_prod_prediction })
sub_RF = sub_RF[['inst_id', 'OC']]

print("데이터 학습 - (1) Random Forest 변수 설정 및 학습 및 테스트 완료")
```

```python
############################################################################
############ GBM
############################################################################
# 11. 데이터 학습 - (2) GBM 변수 설정 및 학습 및 테스트

estimators = 10
np.random.seed(100)
GBM_prod = GradientBoostingClassifier(n_estimators = estimators)

# GBM 훈련 데이터 학습
GBM_prod_model = GBM_prod.fit(train_prod_X, train_prod_Y) 

# GBM 테스트 데이터 테스트
GBM_prod_prediction = GBM_prod.predict_proba(test_prod_X)[:,1]

sub_GBM = pd.DataFrame({'inst_id' : sub_id , 'OC' : GBM_prod_prediction })
sub_GBM = sub_GBM[['inst_id', 'OC']]

print("데이터 학습 - (2) GBM 변수 설정 및 학습 및 테스트 완료")
```

```python
############################################################################
############ XGBOOST
############################################################################
# 12. 데이터 학습 - (3) XGBOOST 변수 설정 및 학습 및 테스트
dtrain_prod = xgb.DMatrix(data = train_prod_X, label = train_prod_Y)
dtest_prod = xgb.DMatrix(data = test_prod_X)

# Custom error function for the XGB model
threshold = 0.5
def eval_error(preds, dtrain):
    labels = dtrain.get_label()
    preds = (preds > threshold ).astype('float')
    return "accuracy", accuracy_score(labels, preds)
    
param = {'objective' : 'binary:logistic',
         'max_depth' : 6,
         'eta': 0.3,
         'colsample_bytree' : 1,
         'subsample' : 1,
         # 'silent' : 0 -> not be used error
         }

nrounds = 2

np.random.seed(100)

# xgb_model 학습 
xgb_model = xgb.train(param, 
                      dtrain_prod, 
                      num_boost_round = nrounds ,
                      feval = eval_error,
                      #maximize = True,
                      #early_stopping_rounds = 10,
                      )

XGB_prediction = xgb_model.predict(dtest_prod)

sub_XGB= pd.DataFrame({'inst_id' : sub_id , 'OC' : XGB_prediction })
sub_XGB= sub_XGB[['inst_id', 'OC']]
  
print("데이터 학습 - (3) XGBOOST 변수 설정 및 학습 및 테스트 완료")
```

```python
# 13. 데이터 학습 - (4) XGBOOST 학습결과 시각화

#PLotting the feature importance
xgb_Imp = pd.DataFrame({'Features' : list(xgb_model.get_score().keys()), 
                        'Importance' : list(xgb_model.get_score().values())}).sort_values(['Importance'])

plt.figure()
sns.barplot(xgb_Imp.Importance, xgb_Imp.Features)
plt.show()
```

```python
############################################################################
#Ensembling the three models
############################################################################

# 14. 데이터 앙상블 - (1) 각 기법 훈련 결과 pandas로 통합 저장
#Forming the ensemble dataset of the 3 models
ensemble = pd.DataFrame()
ensemble['inst_id'] = sub_XGB['inst_id']
ensemble['XGB'] = sub_XGB['OC']
ensemble['GBM'] = sub_GBM['OC']
ensemble['RF'] = sub_RF['OC']

# Taking the average of all 3 models
ensemble['ens'] = (ensemble['XGB'] + ensemble['GBM'] + ensemble['RF'])/3
ensemble['OC'] = (ensemble['ens'] > 0.7).astype('int') #0.7 as the threshold and above that the hospital is closed.
ensemble.head()
```
