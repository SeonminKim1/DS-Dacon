## Dacon - 퇴근시간 버스 승차 인원 예측 (Jeju Island Bus Passengers Estimation in Leaving work time)

### 데이터 설명
#### train, test , bus_bts 공통 사항
- 해당 데이터에는 버스카드를 통해 결제를 한 경우에 대한 정류소 승, 하차 데이터로 모든 승차정보의 경우는 기록이 되어있지만,
- 버스에서 하차를 할 때, 버스카드를 찍지 않는 경우, 해당 기록이 비어 있는 상태입니다. 따라서, 승차 인원수와 하차 인원수가 동일하지 않고 다소 차이가 있음을 미리 알려드립니다.

#### train, test csv 공통사항
- 해당 버스정류장에 대한 각각의 위도, 경도가 제공이 되어있는 상태로 같은 정류장 이름이지만 위도와 경도가 서로 다른 경우가 존재합니다. 
- 해당 경우는, 같은 정류장 이름을 가지고 있는 길 건너편의 정류장에 해당이 됩니다.

#### train.csv and test.csv
- train.csv 의 경우, 2019년 9월 제주도의 각 날짜, 출근시간 (6시~12시)의 버스 정류장별 승하차 인원,  퇴근시간(18시~20시)의 버스 정류장별 승차 인원이 기록되어 있습니다.
- test.csv의 경우, 2019년 10월의 각 날짜, 출근시간(오전 6시~12시)의 버스 정류장별 승하차 인원이 기록되어 있습니다.

#### bus_bts.csv
- bus_bts의 경우, 버스카드별로 승하차 정보가 기록이 되어있습니다. 해당 데이터는 탑승 시간대가오전 6시부터 12시 사이인 경우만 있습니다.

#### submission_제출양식.csv 
- submission_제출양식의 경우, test data의 ID와 목표변수인 18시~20시 승차 인원로만 이루어져 있습니다.
- 참가자분들께서는 test.csv에서 ID와 예측값을 결합하여, 해당 submission_제출양식.csv 파일형식 처럼 만든 이후,
- 해당 제출 파일 양식 그대로(변수명 포함된 상태) 제출을 해주시면 됩니다

```python
import os
os.chdir("C:\\Users\\urse\\Desktop\\dataset\\Dacon-BusJeju")
         
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

print('데이터 로드 완료')
```

```python
print(train.shape)
train.head()
```

```python
print(test.shape)
test.head()
```

```python
# 1. 데이터 탐색 - (1) dtypes
print('#### train_dtypes\n', train.dtypes)
print('#### test_dtypes\n', test.dtypes)
# -> train data와 test data의 데이터 형식은 일치
# -> date 변수 datetime 변수형으로 변환 필요
```

```python
# 1. 데이터탐색 - (2) object형 unique
print('### train data')
for i in train.columns:
    if train[i].dtypes=='object':
        print(i, train[i].unique(), len(train[i].unique()), '개')

print('\n### test data')
for i in test.columns:
    if test[i].dtypes=='object':
        print(i, test[i].unique(), len(test[i].unique()), '개')

# -> train_data는 9월 1일 ~ 9월 30일 / test_data는 10월 1일 ~ 10월 16일
```

```python
# 1. 데이터 탐색 - (3) in_out 값별 갯수
train_in_out_count = train['in_out'].value_counts()
test_in_out_count = test['in_out'].value_counts()

print('train_data\n', train_in_out_count)
print('\ntest_data\n', test_in_out_count)

# -> 후에 시외를 '1'로 시내를 '0'으로 변환 필요
```

```python
# 2. 데이터전처리 - (1) date 변수 변환 및 요일 항목 생성 (train data)
train['date'] = pd.to_datetime(train['date'])

# date 항목으로부터 으로부터 요일을 뽑아낸 후 data frame에 원 핫 인코딩으로 추가
train['weekday'] = train['date'].dt.weekday # 0 ~ 6 (Monday is 0 and Sunday is 6)
train = pd.get_dummies(train, columns=['weekday'])

train.head()
```

```python
# 2. 데이터전처리 - (1) date 변수 변환 및 요일 항목 생성 (test data)
test['date'] = pd.to_datetime(test['date'])

# date 항목으로부터 으로부터 요일을 뽑아낸 후 data frame에 원 핫 인코딩으로 추가
test['weekday'] = test['date'].dt.weekday # 0 ~ 6 (Monday is 0 and Sunday is 6)
test = pd.get_dummies(test, columns=['weekday'])

test.head()
```

```python
# 2. 데이터 전처리 - (2) in_out 변수 변환 -> 
train['in_out'] = train['in_out'].map({'시내':0,'시외':1})
test['in_out'] = test['in_out'].map({'시내':0,'시외':1})
print('in_out 변수 변환 완료')
```

```python
# 3. 데이터 분석 - (1) 좌표 데이터를 이용한 Feature Engineering - 제주도와 서귀포와의 거리 추가

# 제주도의 인구는 서귀포시와 제주시에 몰려 있는 경우가 많아서 해당 지역과의 거리를 각각 feature로 추가할 수 있다.
# dis_jeiusi (버스정류장과 제주시와의 거리) / dis_seoquipo(버스정류장과 서귀포시와의 거리)
# 제주시 위경도. 33.500770, 126.522761
# 서귀포시의 위 경도 : 33.259429, 126.558217

import geopy.distance
coords_jejusi = (33.500770, 126.522761) # 제주시 위경도
coords_seoquipo = (33.259429, 126.558217) # 서귀포시의 위경도

# 해당 버스정류장과 제주시와의 거리를 변수로 넣음
train['dis_jejusi'] = [geopy.distance.vincenty((train['latitude'].iloc[i], train['longitude'].iloc[i]), coords_jejusi).km for i in range(len(train))]
train['dis_seoquipo'] = [geopy.distance.vincenty((train['latitude'].iloc[i], train['longitude'].iloc[i]), coords_seoquipo).km for i in range(len(train))]

print('해당 버스정류장과 제주시와의 거리, 서귀포시와의 거리 추가 완료')
# 제주도와 서귀포 거리 변수 추가
```

```python
# 3. 데이터 분석 - (2) 제주시와의 거리, 서귀포시와의 거리 변수와 18-20 ride 변수 상관관계 구해보기
print('제주시와의 상관관계', train['dis_jejusi'].corr(train['18~20_ride']))
print('서귀포시와의 상관관계', train['dis_seoquipo'].corr(train['18~20_ride']))

# -> 제주시와는 거리가 멀수록 18-20 ride 변수 승차인원이 줄어드는 경향
```

```python
test['dis_jejusi'] = [geopy.distance.vincenty((test['latitude'].iloc[i],test['longitude'].iloc[i]), coords_jejusi).km for i in range(len(test))]
test['dis_seoquipo'] = [geopy.distance.vincenty((test['latitude'].iloc[i],test['longitude'].iloc[i]), coords_seoquipo).km for i in range(len(test))]
print('해당 버스정류장과 제주시와의 거리, 서귀포시와의 거리 추가 완료')

print('제주시와의 상관관계', train['dis_jejusi'].corr(train['18~20_ride']))
print('서귀포시와의 상관관계', train['dis_seoquipo'].corr(train['18~20_ride']))

# -> 제주시와는 거리가 멀수록 18-20 ride 변수 승차인원이 줄어드는 경향
```

```python
# 3. 데이터 분석 - (2) random forest를 활용한 모델링 - 훈련 변수, 타겟 변수 지정 

# 훈련 변수
# id, date, bus_route_id, station_code, station_name 제외됨.
input_var= ['in_out','latitude','longitude','6~7_ride', '7~8_ride', '8~9_ride', '9~10_ride',
       '10~11_ride', '11~12_ride', '6~7_takeoff', '7~8_takeoff', '8~9_takeoff',
       '9~10_takeoff', '10~11_takeoff', '11~12_takeoff','weekday_0', 'weekday_1', 'weekday_2', 'weekday_3', 'weekday_4',
       'weekday_5', 'weekday_6', 'dis_jejusi', 'dis_seoquipo']

# 타겟변수, 목표변수, 종속변수
target=['18~20_ride']

X_train = train[input_var]
y_train = train[target]

X_test = test[input_var]
print('훈련 변수 타겟 변수 지정 완료')
```

```python
# 3. 데이터 분석 - (3) random forest 학습 (시간 좀 걸림)
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=1217)
rf.fit(X_train,y_train)
```

```python
# 4. 테스트 데이터 예측 및 csv저장
test['18~20_ride'] = rf.predict(X_test)
test[['id','18~20_ride']].to_csv("dacon_base_line.csv",index=False)
```
